```octave
function func = f(x)
    func = sin(x)./x.^3;
end
```

```octave
x   = 4;
e   = linspace(1,16);
h   = 10.^(-e);
df  = (x^3*cos(x)-3*x^2*sin(x))/x^6;
df1 = (f(x+h)-f(x))./h;
df2 = (f(x)-f(x-h))./h;
df3 = 0.5*(f(x+h)-f(x-h))./h;
df4 = (f(x-2*h)-8*f(x-h)+8*f(x+h)-f(x+2*h))./12./h;
e1  = abs(df1-df);
e2  = abs(df2-df);
e3  = abs(df3-df);
e4  = abs(df4-df);
plot(-log(h),log(e1),'r')
hold on
plot(-log(h),log(e2),'b')
hold on
plot(-log(h),log(e3),'g')
hold on
plot(-log(h),log(e4),'k')
```
